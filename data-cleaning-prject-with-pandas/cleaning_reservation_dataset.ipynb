```python
import pandas as pd
```

```python
df_hpg_res = pd.read_csv("input/hpg_reserve.csv")
df_hpg_store = pd.read_csv("input/hpg_store_info.csv")
df_date_info = pd.read_csv("input/date_info.csv")
```

```python
df_hpg_res.head()
```

```python
df_hpg_res.dtypes
```

```python
df_hpg_res["visit_datetime"] = pd.to_datetime(df_hpg_res["visit_datetime"])
df_hpg_res["reserve_datetime"] = pd.to_datetime(df_hpg_res["reserve_datetime"])
df_hpg_res["visit_date"] = df_hpg_res["visit_datetime"].dt.date
df_hpg_res["reserve_date"] = df_hpg_res["reserve_datetime"].dt.date
df_hpg_res["visit_date_yyyymm"] = df_hpg_res["visit_datetime"].dt.strftime('%Y%m')
df_hpg_res["reserve_date_yyyymm"] = df_hpg_res["reserve_datetime"].dt.strftime('%Y%m')
```

```python
df_hpg_res.dtypes
```

```python
df_hpg_res.head()
```

```python
df_hpg_store.head()
```

```python
df_hpg_store["prefecture"] = df_hpg_store["hpg_area_name"].str.split(pat=" ", expand=True)[0]
df_hpg_store["city"] = df_hpg_store["hpg_area_name"].str.split(pat=" ", expand=True)[1]
```

```python
df_hpg_store.head()
```

```python
df = df_hpg_res.merge(df_hpg_store, how="inner", on="hpg_store_id")
```

```python
df.head()
```

```python
df_date_info.dtypes
```

```python
df_date_info["calendar_date"] = pd.to_datetime(df_date_info["calendar_date"])
df_date_info["calendar_date"] = df_date_info["calendar_date"].dt.date
```

```python
df = df.merge(df_date_info, how="inner", left_on="visit_date", right_on="calendar_date")
df.rename(columns={"day_of_week": "day_of_week_visit"}, inplace=True)
df.rename(columns={"holiday_flg": "holiday_flg_visit"}, inplace=True)
```

```python
df = df.merge(df_date_info, how="inner", left_on="reserve_date", right_on="calendar_date")
df.rename(columns={"day_of_week": "day_of_week_reserve"}, inplace=True)
df.rename(columns={"holiday_flg": "holiday_flg_reserve"}, inplace=True)
```

```python
df = df.drop(columns=["calendar_date_x", "calendar_date_y"])
```
